## 2. 数据清理/Data Cleaning

### a. 定义方程 / define functions

#### 为数据科学专栏加入难度评级 / pair the difficulty accordig each 

```python
def merge_two_df(df1,df2):
    df2 = df2.merge(df1, how = 'left', left_on='题号', right_on='#')
    del df2['Title']
    del df2['Solution'] 
    del df2['#']
    return df2
```

#### 删除各个分类的无用列，删除次级分类，将题号转为整数型 / delete useless columns and sub-category from dictionary, convert # as int 

```python
def clean_data(dictionary):
    lst_of_name = list(dictionary.keys())
    
    
    for i in range(len(lst_of_name)):
        if 'Unnamed: 3' in dictionary[lst_of_name[i]]:
            del dictionary[lst_of_name[i]]['Unnamed: 3']
            
        elif '说明' in dictionary[lst_of_name[i]]:
            del dictionary[lst_of_name[i]]['说明']
            
        dictionary[lst_of_name[i]] = dictionary[lst_of_name[i]].dropna(axis = 0)
        
        
    for i in range(len(lst_of_name)):
        dictionary[lst_of_name[i]]['题号']= dictionary[lst_of_name[i]]['题号'].astype(int)
```

#### 按照题库分类，为数据科学专栏加入分类 / pair the category for each DS-leetcode questions

```python
def Classifier(dictionary,df):
    
    lst_of_name = list(dictionary.keys())
    
    for i in range(len(lst_of_name)):
        for j in range(len(df)):
            if df['题号'][j] in dictionary[lst_of_name[i]]['题号'].values:
                df['讲解链接'][j] = lst_of_name[i]
```

### b. 数据整合/ Data Manipulation

#### 匹配题目难度 / Pair the difficulty

```python
leetcode_for_DS = merge_two_df(leetcode_with_diffculty,leetcode_for_DS)
leetcode_for_DS.head(5)
```

#### 清洗字典/ cleaning the data from the dictionary

```python
clean_data(leetcode)
leetcode
```

#### 按照分类为数据科学专栏题目进行分类 / pair the category for DS leetcode

```python
Classifier(leetcode,leetcode_for_DS)
leetcode_for_DS.head(5)
```

#### 重命名，删除无用列 / rename,reindex,dropna

```python
leetcode_for_DS = leetcode_for_DS.dropna(axis =1)
leetcode_for_DS.columns = ['#','Title','Category','Difficulty']
leetcode_for_DS.set_index(leetcode_for_DS['#'],inplace=True)
leetcode_for_DS.drop(columns = '#',inplace=True)
leetcode_for_DS.head(10)
```

## 3.数据整合 / Data integration

#### 按照分类放入字典 / split each category and compress them into dictionary

```python
def retrun_dic_DataFrame(df):
    leetcode_for_dic = {i:df[df['Category']==i] for i in df['Category'].unique()}
    return leetcode_for_dic
```
